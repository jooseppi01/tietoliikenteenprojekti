```python
from sklearn.metrics import confusion_matrix
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
#data tallennettu arduinolta putty2.log nimiseen tiedostoon.
data = np.loadtxt("putty2.log")
y_test = data[:, 0]
y_pred = data[:, 1]

kk = confusion_matrix(y_test, y_pred)
display3 = metrics.ConfusionMatrixDisplay(confusion_matrix = kk, display_labels = ['p1', 'p2', 'p3', 'p4'])
display3.plot()
```

```python
#data tallennettu arduinolta putty3.log nimiseen tiedostoon, testailua hieman isommalla datalla.
data = np.loadtxt("putty3.log")

y_test = data[:, 0]
y_pred = data[:, 1]

kk = confusion_matrix(y_test, y_pred)
display3 = metrics.ConfusionMatrixDisplay(confusion_matrix = kk, display_labels = ['p1', 'p2', 'p3', 'p4'])
display3.plot()
```

```python
score = metrics.accuracy_score(y_test, y_pred)
print('tarkkuus:', score)
```
